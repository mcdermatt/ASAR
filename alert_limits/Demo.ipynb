## Rotation Matrices
Let's take a minute to visualize what's going on when we use a rotation matrix

```python
from utils import rot

#try adjusting these two parameters and seeing what happens-------------------
angle = -60
point = np.array([3,4])
#-----------------------------------------------------------------------------

R = rot(angle)
print("rotation matrix R: \n", R)
print("\n Origonal point:", point)
transformed = R.dot(point)
print("\n Transformed by R:", transformed)

fig, ax = plt.subplots()
ax.set_xlim([-6,6])
ax.set_ylim([-6,6])
ax.set_aspect("equal")
ax.plot(point[0],point[1], 'r.', ms = 20)
ax.plot([0,point[0]], [0,point[1]], 'k--')
ax.plot(transformed[0],transformed[1], 'b.', ms = 20)
ax.plot([0,transformed[0]], [0,transformed[1]], 'k--')
```
