```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# for auto-reloading external modules
%load_ext autoreload
%autoreload 2
%autosave 180
# %matplotlib notebook
```

```python
from p_loop import p_loop

n = 30#number of cycles to calculate ss
# m = 100 #number of cycles to display

x_hist, y_hist, theta_hist = p_loop(niter = n)

fig1, ax1 = plt.subplots(3)
fig1.tight_layout(h_pad = 5)

ax1[0].plot(x_hist)
ax1[0].set_ylabel("_x (m)")
ax1[0].set_xlabel("time (s)")
strx = "_x steady state = " + str(round(np.min(x_hist), 2)) + " m"
ss1 = np.linspace(0,n-1,10)
ax1[0].plot(ss1, np.min(x_hist)*np.ones(10), "r--", label = strx)
ax1[0].legend()

ax1[1].plot(y_hist)
ax1[1].set_ylabel("_y (m)")
ax1[1].set_xlabel("time (s)")
stry = "_y steady state = " + str(round(np.min(y_hist), 2)) + " m" 
ss1 = np.linspace(0,n-1,10)
ax1[1].plot(ss1, np.min(y_hist)*np.ones(10), "r--", label = stry)
ax1[1].legend()

ax1[2].plot(theta_hist)
ax1[2].set_ylabel("_theta (rad)")
ax1[2].set_xlabel("time (s)")
strtheta = "_theta steady state = " + str(round(np.min(theta_hist),2)) + " rad"
ss2 = np.linspace(0,n-1,10)
ax1[2].plot(ss1, np.min(theta_hist)*np.ones(10), "r--", label = strtheta)
ax1[2].legend()

plt.show()
```
